```
# Étape 1 : Importer les bibliothèques
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Étape 2 : Charger le jeu de données (assurez-vous que le fichier est bien dans votre environnement Colab)
df = pd.read_csv("Employee Attrition.csv")

# Étape 3 : Aperçu rapide
df.head()


# Mesures de tendance centrale
       # Moyennes
print("Moyenne d'âge :", df["Age"].mean())
print("Moyenne du salaire mensuel :", df["MonthlyIncome"].mean())

       # Médianes
print("Médiane d'âge :", df["Age"].median())
print("Médiane du salaire mensuel :", df["MonthlyIncome"].median())

       # Modes
print("Mode d'âge :", df["Age"].mode()[0])
print("Mode du salaire mensuel :", df["MonthlyIncome"].mode()[0])


# Mesures de dispersion

      # Variance
print("Variance d'âge :", df["Age"].var())
print("Variance du salaire mensuel :", df["MonthlyIncome"].var())

      # Écart-type
print("Écart-type d'âge :", df["Age"].std())
print("Écart-type du salaire mensuel :", df["MonthlyIncome"].std())

      # Étendue (valeur max - min)
print("Étendue d'âge :", df["Age"].max() - df["Age"].min())
print("Étendue du salaire mensuel :", df["MonthlyIncome"].max() - df["MonthlyIncome"].min())



# Visualisations

     # Histogramme de l'âge
plt.figure(figsize=(8, 4))
sns.histplot(df["Age"], bins=20, kde=True, color="skyblue")
plt.title("Distribution de l'âge des employés")
plt.xlabel("Âge")
plt.ylabel("Nombre d'employés")
plt.grid(True)
plt.show()

     # Histogramme du salaire
plt.figure(figsize=(8, 4))
sns.histplot(df["MonthlyIncome"], bins=30, kde=True, color="salmon")
plt.title("Distribution du salaire mensuel")
plt.xlabel("Salaire mensuel")
plt.ylabel("Nombre d'employés")
plt.grid(True)
plt.show()


     # Boxplot de l'âge
plt.figure(figsize=(6, 4))
sns.boxplot(x=df["Age"], color="lightgreen")
plt.title("Boxplot de l'âge")
plt.xlabel("Âge")
plt.grid(True)
plt.show()

     # Boxplot du salaire
plt.figure(figsize=(6, 4))
sns.boxplot(x=df["MonthlyIncome"], color="orange")
plt.title("Boxplot du salaire mensuel")
plt.xlabel("Salaire mensuel")
plt.grid(True)
plt.show()



```

## Introduction
Ce mini-projet vise à analyser un jeu de données sur des employés, en mettant en évidence les mesures de tendance centrale (moyenne, médiane, mode) et de dispersion (variance, écart-type, étendue).

Les données analysées proviennent d’un fichier Kaggle portant sur l’attrition des employés.




## Mesures de tendance centrale

Les résultats obtenus sont :

- **Âge moyen** : 36.923809523809524 ans
- **Âge médian** : 36.0 ans
- **Âge mode** : 35 ans

- **Salaire mensuel moyen** : 6502.931292517007 $

- **Salaire mensuel médian** : 4919.0 $

- **Salaire mensuel mode** : 2342 $

### Interprétation
L’âge moyen et médian étant proches, cela indique une distribution relativement symétrique. Le salaire, quant à lui, montre une certaine asymétrie avec un mode plus bas que la moyenne.



## Mesures de dispersion

- **Écart-type âge** : 9.135373489136734
- **Variance âge** : 83.45504878602227
- **Étendue âge** : 42

- **Écart-type salaire** : 4707.956783097995
- **Variance salaire** : 22164857.07151842
- **Étendue salaire** : 18990

### Interprétation
Une forte dispersion du salaire est visible, ce qui montre une grande hétérogénéité entre les employés. L’âge est plus concentré autour de la moyenne.





## Visualisations

Les histogrammes et boxplots permettent de mieux visualiser les répartitions :

- L’histogramme d’âge montre une concentration autour de 30-40 ans.
- Le boxplot du salaire indique la présence de valeurs extrêmes (outliers), probablement des hauts salaires rares.
